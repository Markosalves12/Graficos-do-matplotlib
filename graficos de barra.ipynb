```python
#vetor com os labels x
x = ['python', 'c#', 'Java', 'Dart']
#quantidades respectivas
y = [51,62,63,54]

#plotagem das barras individualmente
#o primeiro parametro e sempre x, o segundo o y
fig, ax = plt.subplots(figsize=(14,8))
plt.bar(x[0], y[0], label=x[0], color = 'green')
plt.bar(x[1], y[1], label=x[1], color = 'yellow')
plt.bar(x[2], y[2], label=x[2], color = 'blue')
plt.bar(x[3], y[3], label=x[3], color = 'red')

#configuração e estilos do titulo
plt.title("demonstração", fontdict  = {'fontsize': '15',
                                       'family': 'serif',
                                       'fontweight': 'bold',
                                       'color': 'black' }, loc = 'center', )
#configuração e estilos do label x
plt.xlabel("Linguagem d eprogramação", fontdict  = {'fontsize': '15',
                                       'family': 'serif',
                                       'fontweight': 'bold',
                                       'color': 'blue' }, loc = 'center', )
#configuração e estilos do label y
plt.ylabel("pontuação", fontdict  = {'fontsize': '15',
                                       'family': 'serif',
                                       'fontweight': 'bold',
                                       'color': 'blue', 'verticalalignment': 'baseline',
                                        }, loc = 'center')
#legenda automatica configuração e posicionamento da 
ax.legend(loc = 'upper right')
plt.savefig("C:/Users/Markos Alves/Downloads/MATPLOTLIB/fig.jpeg", transparent = False, edgecolor =  'white', )
plt.show()
```
